### Importing Important Libraries

```python
import numpy as np
import pandas as pd
```

### Exploring the data

```python
df = pd.read_csv("Mall_Customers.csv")
df.head(n=10)
```

### Checking size of data

```python
df.shape
```

There are 200 customers with 5 feature in data.

### Checking data types of columns

```python
df.info()
```

### Renaming the columns

```python
df = df.rename(columns={'Annual Income (k$)': 'Annual_income', 'Spending Score (1-100)': 'Spending_score','Genre':'Gender'})
```

```python
df.head()
```

### Removing CustomerId column

Since this feature has no impact on beaviour of customers, we can discard this column

```python
df.drop('CustomerID',axis=1,inplace=True)
df.head()
```

```python
df['Gender'].replace(['Female','Male'], [0,1],inplace=True)
```

### Checking Null values

```python
df.isnull().sum()
```

If any of the sum is non zero we can remove null values or replace null values with mean depending on amount of null data.

# Visualizations

We will now try to get better insight of data using matplotlib and seaborn library.

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
```

#### Counting the number of Males and Females

```python
sns.countplot(x='Gender',data=df,palette="twilight_r")
```

```python
plt.figure(1 , figsize = (15 , 6))
feature_list = ['Age','Annual_income', "Spending_score"]
feature_listt = ['Age','Annual_income', "Spending_score"]
pos = 1 
for i in feature_list:
    plt.subplot(1 , 3 , pos)
    plt.subplots_adjust(hspace = 0.5 , wspace = 0.5)
    sns.distplot(df[i], bins=20, kde = True,color="Green")
    pos = pos + 1
plt.show()
```

```python
sns.pairplot(df,  kind ="reg", hue = "Gender", palette="twilight")
plt.tight_layout()
```

### Age VS Spending score

```python
sns.lmplot(x = "Age", y = "Spending_score", data = df, hue = "Gender",palette="twilight")
```

### Annual Income VS Spending score

```python
sns.lmplot(x = "Annual_income", y = "Spending_score", data = df, hue = "Gender",palette="twilight")
```

```python
sns.lmplot(y = "Annual_income", x = "Age", data = df, hue = "Gender",palette="twilight")
```

```python
sns.stripplot(x="Gender", y="Spending_score", data=df)
plt.tight_layout()
```
